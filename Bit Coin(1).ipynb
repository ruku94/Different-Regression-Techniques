## Ridge

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

grid_search = GridSearchCV(ridge, param_grid, cv=10)

grid_search.fit(X, y)
```
